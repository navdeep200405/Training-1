```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)


fig, axs = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(12, 6),              
    dpi=100,                      
    sharex='col',                 
    sharey='row',                
    facecolor='lightgray',     
    constrained_layout=True     
)

axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('sin(x)')

axs[0, 1].plot(x, y2, color='green')
axs[0, 1].set_title('cos(x)')

axs[1, 0].plot(x, y3, color='red')
axs[1, 0].set_title('tan(x)')
axs[1, 0].set_ylim(-5, 5)  # Limit y-axis for better view

axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title('exp(-x)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt  

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 6), dpi=100)


axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title("sin(x)")
axs[0, 0].grid(True)


axs[0, 1].plot(x, y2, color='green')
axs[0, 1].set_title("cos(x)")
axs[0, 1].grid(True)


axs[1, 0].plot(x, y3, color='red')
axs[1, 0].set_title("tan(x)")
axs[1, 0].set_ylim(-10, 10)  
axs[1, 0].grid(True)

axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_title("exp(x)")
axs[1, 1].grid(True)

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/titanic-train1-csv/titanic_train.csv")
df.columns = df.columns.str.strip().str.lower()

print(df.columns)

parch = df['parch'].value_counts().sort_index()
parch.plot(kind="bar", figsize=(10, 8), color='r')
plt.ylabel("Number of Passengers")
plt.xlabel("Number of Parents/Children")
plt.title("Number of Passengers Travelling with their Parents/Children")
plt.show()
```

```python
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Mid-Age', 'Senior']
df_titanic['AgeGroup'] = pd.cut(df_titanic['Age'], bins=bins, labels=labels)
age_group_survival = df_titanic.groupby('AgeGroup')['Survived'].mean()
age_group_survival.plot(kind='bar')
plt.title('Survival Rate by Age Group')
plt.ylabel('Survival Rate')
plt.show()
```

```python
df_titanic['FamilySize'] = df_titanic['SibSp'] + df_titanic['Parch']
family_survival = df_titanic.groupby('FamilySize')['Survived'].mean()

plt.plot(family_survival.index, family_survival.values, marker='o')
plt.title('Survival Rate by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Survival Rate')
plt.grid()
plt.show()
```
