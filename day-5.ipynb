```python
import pandas as pd

s = pd.Series(['php', 'python', 'java', 'c#'])
print("Original Series:")
print(s)

def convert(word):
    if len(word) > 1:
        return word[0].upper() + word[1:-1] + word[-1].upper()
    else:
        return word.upper()

result = s.apply(convert)

print("\nFirst and last character of each word to upper case:")
print(result)
```

```python
big_wkt_wins = df[(df['result'] == 'wickets') & (df['result_margin'] > 5)]  
print(len(big_wkt_wins))
```

```python
big_run_wins = df[df['result'] == 'runs'].nlargest(10, 'result_margin')  
print(big_run_wins[['winner', 'result_margin', 'season']])
```

```python
season_team_wins = (df.groupby(['season', 'winner'])
                      .size()
                      .unstack(fill_value=0)
                      .sort_index())   

print(season_team_wins)
```

```python
import matplotlib.pyplot as plt

matches_per_season = df['season'].value_counts().sort_index()

matches_per_season.plot(kind='bar')
plt.title('IPL Matches per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.tight_layout()
plt.show()
```

```python
played = pd.Series(df['team1'].tolist() + df['team2'].tolist()).value_counts()  
wins = df['winner'].value_counts()  

win_rate = (wins / played * 100).round(2).sort_values(ascending=False)   
print(win_rate.dropna())   
```

```python
played = pd.Series(df['team1'].tolist() + df['team2'].tolist()).value_counts()   
wins = df['winner'].value_counts()  

win_rate = (wins / played * 100).round(2).sort_values(ascending=False)  
print(win_rate.dropna()) 
```

```python
toss_trend = (df.groupby(['season', 'toss_decision'])
                 .size()
                 .unstack(fill_value=0)
                 .apply(lambda s: s/s.sum()*100, axis=1))   # % base toss decisions

toss_trend.plot()
plt.title('Toss Decision % by Season')
plt.ylabel('% of Matches')
plt.show()
```

```python
top10 = df['player_of_match'].value_counts().head(10)   

plt.figure()                           
top10.plot(kind='bar')
plt.title('Top 10 Players – Player of Match Awards')   
plt.xlabel('Player')
plt.ylabel('Awards Count')
plt.xticks(rotation=45, ha='right')     
plt.tight_layout()
plt.show()
```

```python
player = 'V Kohli'                      

season_pom = (df[df['player_of_match'] == player]
                .groupby('season')
                .size()
                .sort_index())

plt.figure()
season_pom.plot(marker='o')             
plt.title(f'{player} – Awards per Season')
plt.xlabel('Season')
plt.ylabel('Number of Awards')
plt.grid(True)                           
plt.tight_layout()
plt.show()
```

```python
finals = df.groupby('season').tail(1)   

cross = pd.crosstab(finals['winner'], finals['player_of_match'])  
cross = cross.loc[cross.sum(axis=1).sort_values(ascending=False).index]  

plt.figure()
cross.plot(kind='bar', stacked=True)
plt.title('Finals – Winning Team vs Player of Match')
plt.xlabel('Winning Team')
plt.ylabel('Count (should be 1 per season)')   
plt.tight_layout()
plt.show()
```

```python
same_player = df['toss_winner'] == df['player_of_match']  
share = same_player.value_counts()

plt.figure()
share.plot(kind='pie', labels=['Yes', 'No'], autopct='%1.1f%%', startangle=90)
plt.title('Did Toss Winner also become Player of Match?')
plt.ylabel('')       
plt.tight_layout()
plt.show()
```

```python
player = 'MS Dhoni'                              

runs_df = df[(df['player_of_match'] == player) & (df['result'] == 'runs')]

plt.figure()
plt.hist(runs_df['result_margin'])               
plt.title(f'{player} – Run Margins when Player of Match')
plt.xlabel('Runs Margin')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/matches/matches.csv")   

top_n = 10
counts = df['player_of_match'].value_counts()
top_players = counts.head(top_n)
top_players['Others'] = counts.iloc[top_n:].sum()

plt.figure(figsize=(8, 8))
top_players.plot(kind='pie',
                 autopct='%1.1f%%',
                 startangle=90,
                 labels=top_players.index)
plt.title('Top 10 Players + Others – Player of the Match Awards')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
