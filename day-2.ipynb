```python
num = int(input("Enter a number: "))
power = len(str(num))  # Number of digits
sum = 0
temp = num

while temp > 0:
    digit = temp % 10
    sum += digit ** power
    temp = temp // 10

if sum == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```
