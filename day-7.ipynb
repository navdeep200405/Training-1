```python
import re

def is_valid_string(s):
    return bool(re.fullmatch(r'[a-zA-Z0-9]+', s))

print(is_valid_string("abc123"))  # ✅ True
print(is_valid_string("abc-123")) # ❌ False
```

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['John', 'Samantha', 'Alex', 'Nina', 'Mark', 'Lilly', 'Tom', 'Olivia'],
    'Salary': [50000, 62000, np.nan, 58000, 54000, np.nan, 60000, 61000]
}

df = pd.DataFrame(data)

df_ffill = df.copy()
df_ffill['Salary'] = df_ffill['Salary'].fillna(method='ffill')

print("\nForward Fill:\n")
print(df_ffill[['Name', 'Salary']])
```

```python
import pandas as pd
import numpy as np

data = {
    'Employee': ['Aman', 'Bina', 'Charan', 'Deepa', 'Esha'],
    'Age': [25, np.nan, 30, np.nan, 28],
    'Salary': [50000, 60000, np.nan, 58000, np.nan],
    'Department': ['HR', np.nan, 'IT', 'Finance', np.nan]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Mean Imputation
df_mean = df.copy()
df_mean['Age'] = df_mean['Age'].fillna(df_mean['Age'].mean())
df_mean['Salary'] = df_mean['Salary'].fillna(df_mean['Salary'].mean())
print("\nAfter Mean Imputation:\n", df_mean)

# Median Imputation
df_median = df.copy()
df_median['Age'] = df_median['Age'].fillna(df_median['Age'].median())
df_median['Salary'] = df_median['Salary'].fillna(df_median['Salary'].median())
print("\nAfter Median Imputation:\n", df_median)

# Mode Imputation
df_mode = df.copy()
df_mode['Department'] = df_mode['Department'].fillna(df_mode['Department'].mode()[0])
print("\nAfter Mode Imputation (Department):\n", df_mode)
```

```python
import pandas as pd
import numpy as np

data = {
    'Employee': ['Aman', 'Bina', 'Charan', 'Deepa', 'Esha'],
    'Salary': [50000, 60000, np.nan, 58000, np.nan]
}

df = pd.DataFrame(data)

df_bfill = df.copy()
df_bfill['Salary'] = df_bfill['Salary'].fillna(method='bfill')

print("\nBackward Fill:\n")
print(df_bfill[['Employee', 'Salary']])
```
