```python
rows = int(input('Enter rows: '))
for i in range(1, rows+1):
    for j in range(1, i+1):
        print(j, end="")
    for k in range(i-1, 0, -1):
        print(k, end="")
print()
```

```python
rows = int(input("Enter rows: "))
for i in range(rows, 0, -1):
# spaces
    for s in range(rows - i):
        print(" ", end="")
# ascending
    for j in range(1, i + 1):
        print(j, end="")
# descending
    for k in range(i - 1, 0, -1):
        print(k, end="")
print()
```
